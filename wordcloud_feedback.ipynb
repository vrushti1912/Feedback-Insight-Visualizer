```python
# If you would like to work with the raw text you can use 'moby_raw' with open('moby.txt', 'r') as f:
import json
with open('/content/drive/MyDrive/hack/reviews.json', 'r') as f:
  data = json.load(f)
review_text = [x["review_text"] for x in data]
print(review_text)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
tokenized_reviews = [word_tokenize(review) for review in review_text]
print(tokenized_reviews)
```

```python
tokenised_reviews = [review.split() for review in review_text]
unique_tokens = set()
for review in tokenised_reviews:
  for token in review:
    unique_tokens.add(token)
print(unique_tokens)
```

```python
from collections import Counter

# Create a list of all the tokens in the review_text list
tokens = [word for review in review_text for word in review.split()]

# Count the frequency of each unique token
token_counts = Counter(tokens)

d={}
# Print the frequency of each unique token
for token, count in token_counts.items():
    if count > 0:
        print(f"{token}: {count}")
        d[token]=count

# Store tokens with frequency > 14 in a list named 'topic'
topic =[token for token, count in token_counts.items() if count > 0]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag

stop_words = stopwords.words('english')
articles = ['a', 'an', 'the','The','A','An']
pronouns = ['i', 'you', 'he', 'she', 'it', 'we', 'they','I','we','It','it',"It's"]
positive_words = ['good', 'great', 'excellent', 'awesome', 'amazing','like','love']
negative_words = ['bad', 'terrible', 'horrible', 'awful', 'disgusting','worse','worst']
verbs = []
adverbs = []


for word, pos in pos_tag(topic):
    if pos.startswith('VB'):
        verbs.append(word)
    if pos.startswith('RB'):
        adverbs.append(word)

final = [word for word in topic if word not in stop_words and word not in articles and word not in pronouns and word not in positive_words and word not in negative_words and word not in verbs and word not in adverbs]

print(final)
```

```python
import re
#final = [x for x in final if x]
final = [x.replace(',', '').replace(' ', '').replace('.', '').replace('"', '') for x in final]
#final = [re.sub( r"\W+","", x) for x in final]
print(list(set(final)))
```
